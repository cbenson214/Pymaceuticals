```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "Mouse_metadata.csv"
study_results_path = "Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
data = pd.merge(study_results,mouse_metadata, on='Mouse ID', how='outer')

# Display the data table for preview
data.head()
```

```python
# Checking the number of mice.
len(data["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dup_rows = data[data.duplicated(['Mouse ID', 'Timepoint'])]
dup_mouse = dup_rows.iloc[0,0]
dup_mouse
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_data = data.loc[data['Mouse ID'] == dup_mouse, :]
dup_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
data = data.set_index('Mouse ID')
data = data.drop(index=dup_mouse)
data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
data.reset_index(level='Mouse ID', inplace = True)
len(data["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
drug_df = data[['Drug Regimen', 'Tumor Volume (mm3)']]

# mean, median, variance, standard deviation, and SEM of the tumor volume. 
drug_mean_df = drug_df.groupby(['Drug Regimen']).mean()
drug_median_df = drug_df.groupby(['Drug Regimen']).median()
drug_var_df = drug_df.groupby(['Drug Regimen']).var()
drug_std_df = drug_df.groupby(['Drug Regimen']).std()
drug_sem_df = drug_df.groupby(['Drug Regimen']).sem()

# Assemble the resulting series into a single summary DataFrame.
drug_summary = pd.merge(drug_mean_df, drug_median_df, on='Drug Regimen', how='outer')
drug_summary = drug_summary.rename(columns={'Tumor Volume (mm3)_x' : 'Mean Tumor Volume', 'Tumor Volume (mm3)_y' : 'Median Tumor Volume'})
drug_summary = pd.merge(drug_summary, drug_var_df, on='Drug Regimen', how='outer')
drug_summary = pd.merge(drug_summary, drug_std_df, on='Drug Regimen', how='outer')
drug_summary = pd.merge(drug_summary, drug_sem_df, on='Drug Regimen', how='outer')
drug_summary = drug_summary.rename(columns={'Tumor Volume (mm3)_x' : 'Tumor Volume Variance',
                                            'Tumor Volume (mm3)_y' : 'Tumor Std. Dev.',
                                            'Tumor Volume (mm3)' : 'Tumor Std. Err.'})
drug_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
grouped_df = drug_df.groupby('Drug Regimen')
agg_df = grouped_df['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
agg_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
    
# Start by getting the last (greatest) timepoint for each mouse
idx = study_results.groupby('Mouse ID')['Timepoint'].idxmax()
mice = study_results.loc[idx]
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
max_data = pd.merge(mice, mouse_metadata, on='Mouse ID', how='outer')
max_data.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []
mouse = []
drug = []

# Locate the rows which contain mice on each drug and get the tumor volumes
for x in range(max_data['Mouse ID'].count()):
    check = max_data.iloc[x,4]
    if check in treatments:
        tumor_vol.append(max_data.iloc[x,2])
        drug.append(check)
        mouse.append(max_data.iloc[x,0])
new_max = pd.DataFrame(list(zip(drug, mouse, tumor_vol)),
                      columns=['Drug Regimen', 'Mouse ID', 'Tumor Volume'])

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    #Capomulin
capo_max_df = new_max.loc[new_max['Drug Regimen']=='Capomulin']
capo_quartiles = capo_max_df['Tumor Volume'].quantile([.25,.5,.75])
capo_lowerq = capo_quartiles[.25]
capo_upperq = capo_quartiles[.75]
capo_median = capo_quartiles[.5]
capo_iqr = capo_upperq - capo_lowerq
capo_lower_bound = capo_lowerq - (1.5 * capo_iqr)
capo_upper_bound = capo_upperq + (1.5 * capo_iqr)

#Ramicane
ram_max_df = new_max.loc[new_max['Drug Regimen']=='Ramicane']
ram_quartiles = ram_max_df['Tumor Volume'].quantile([.25,.5,.75])
ram_lowerq = ram_quartiles[.25]
ram_upperq = ram_quartiles[.75]
ram_median = ram_quartiles[.5]
ram_iqr = ram_upperq - ram_lowerq
ram_lower_bound = ram_lowerq - (1.5 * ram_iqr)
ram_upper_bound = ram_upperq + (1.5 * ram_iqr)

#Infubinol
inf_max_df = new_max.loc[new_max['Drug Regimen']=='Infubinol']
inf_quartiles = inf_max_df['Tumor Volume'].quantile([.25,.5,.75])
inf_lowerq = inf_quartiles[.25]
inf_upperq = inf_quartiles[.75]
inf_median = inf_quartiles[.5]
inf_iqr = inf_upperq - inf_lowerq
inf_lower_bound = inf_lowerq - (1.5 * inf_iqr)
inf_upper_bound = inf_upperq + (1.5 * inf_iqr)

#Ceftamin
ceft_max_df = new_max.loc[new_max['Drug Regimen']=='Ceftamin']
ceft_quartiles = ceft_max_df['Tumor Volume'].quantile([.25,.5,.75])
ceft_lowerq = ceft_quartiles[.25]
ceft_upperq = ceft_quartiles[.75]
ceft_median = ceft_quartiles[.5]
ceft_iqr = ceft_upperq - ceft_lowerq
ceft_lower_bound = ceft_lowerq - (1.5 * ceft_iqr)
ceft_upper_bound = ceft_upperq + (1.5 * ceft_iqr)

    # add subset 
subset = new_max.sample(20)

    # Determine outliers using upper and lower bounds
capo_outlier = capo_max_df[(capo_max_df['Tumor Volume'] < capo_lower_bound) | (new_max['Tumor Volume'] > capo_upper_bound)]
ram_outlier = ram_max_df[(ram_max_df['Tumor Volume'] < ram_lower_bound) | (new_max['Tumor Volume'] > ram_upper_bound)]
inf_outlier = inf_max_df[(inf_max_df['Tumor Volume'] < inf_lower_bound) | (new_max['Tumor Volume'] > inf_upper_bound)]
ceft_outlier = ceft_max_df[(ceft_max_df['Tumor Volume'] < ceft_lower_bound) | (new_max['Tumor Volume'] > ceft_upper_bound)]
print(f"Capomulin's potential outliers: {capo_outlier}")
print(f"Ramicane's potential outliers: {ram_outlier}")
print(f"Infubinol's potential outliers: {inf_outlier}")
print(f"Ceftamin's potential outliers: {ceft_outlier}")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

    # Correlation Coefficient
filtered_avg_weight_data = avg_weight_data[['Weight (g)','Tumor Volume (mm3)']]
correlation_matrix = filtered_avg_weight_data.corr()
correlation_coefficient = correlation_matrix.iloc[0,1]
correlation_coefficient
```

```python
    # Linear Regression Model
x_values = filtered_avg_weight_data['Weight (g)']
y_values = filtered_avg_weight_data['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
line_eq
```

```python
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
