```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "Mouse_metadata.csv"
study_results_path = "Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
data = pd.merge(study_results,mouse_metadata, on='Mouse ID', how='outer')

# Display the data table for preview
data.head()
```

```python
# Checking the number of mice.
len(data["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dup_rows = data[data.duplicated(['Mouse ID', 'Timepoint'])]
dup_mouse = dup_rows.iloc[0,0]
dup_mouse
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_data = data.loc[data['Mouse ID'] == dup_mouse, :]
dup_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
data = data.set_index('Mouse ID')
data = data.drop(index=dup_mouse)
data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
data.reset_index(level='Mouse ID', inplace = True)
len(data["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
drug_df = data[['Drug Regimen', 'Tumor Volume (mm3)']]

# mean, median, variance, standard deviation, and SEM of the tumor volume. 
drug_mean_df = drug_df.groupby(['Drug Regimen']).mean()
drug_median_df = drug_df.groupby(['Drug Regimen']).median()
drug_var_df = drug_df.groupby(['Drug Regimen']).var()
drug_std_df = drug_df.groupby(['Drug Regimen']).std()
drug_sem_df = drug_df.groupby(['Drug Regimen']).sem()

# Assemble the resulting series into a single summary DataFrame.
drug_summary = pd.merge(drug_mean_df, drug_median_df, on='Drug Regimen', how='outer')
drug_summary = drug_summary.rename(columns={'Tumor Volume (mm3)_x' : 'Mean Tumor Volume', 'Tumor Volume (mm3)_y' : 'Median Tumor Volume'})
drug_summary = pd.merge(drug_summary, drug_var_df, on='Drug Regimen', how='outer')
drug_summary = pd.merge(drug_summary, drug_std_df, on='Drug Regimen', how='outer')
drug_summary = pd.merge(drug_summary, drug_sem_df, on='Drug Regimen', how='outer')
drug_summary = drug_summary.rename(columns={'Tumor Volume (mm3)_x' : 'Tumor Volume Variance',
                                            'Tumor Volume (mm3)_y' : 'Tumor Std. Dev.',
                                            'Tumor Volume (mm3)' : 'Tumor Std. Err.'})
drug_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
grouped_df = drug_df.groupby('Drug Regimen')
agg_df = grouped_df['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
agg_df
```
